```python
import json

data = open('BengaliDictionary.json','r',encoding='utf-8-sig').read()
jdata = json.loads(data)
print(jdata[0])
```

```python
data_list = []
for data in jdata:
    if (data['bn'] is not None) and (data['pron'][-1] is not None):
        data_list.append([data['bn'],data['pron'][-1]])
print(len(data_list))
```

```python
with open('dictionary_extracted_raw.tsv','w',encoding='utf-8') as fop:
    for d in data_list:
        fop.write(d[0]+"\t"+d[1]+"\n")
```

```python
special_chars = set()
for d in data_list:
    for ch in d[-1] :
        if ord(ch) > 128:
            special_chars.add(ch)
print(len(special_chars))
```

```python
with open('special_chars.tsv','w',encoding='utf-8') as fop:
    for ch in special_chars:
        fop.write(ch+'\n')
```

```python
with open('special_chars_mapping.tsv','r',encoding='utf-8-sig') as fip:
    sp_chars = fip.read().splitlines()
```

```python
sp_chars_dict = dict()
for line in sp_chars:
    key_value = line.split('\t')
    print(key_value)
    key = key_value[0]
    value = key_value[1]
    sp_chars_dict[key] = value
```

```python
import string
def discard(s,threshold=0.99):
    if s is None or len(s) < 1:
        return True
    cnt = 0
    for ch in s:
        if ch in string.ascii_letters:
            cnt += 1
    if float(cnt)/len(s) >= threshold:
        return True
    else:
        return False
    
assert(discard('খোলা শেষ') == False)
assert(discard('apple') == True)
```

```python
from collections import defaultdict

data_dict = defaultdict(list)
data_list_processed = []
    
for j,d in enumerate(data_list):
    if discard(d[0],0.5) == True:
        print(d,"discard")
        continue
    str_builder = ''
    for ch in d[1]:
        if ord(ch) > 128:
            if ch in sp_chars_dict:
                str_builder += sp_chars_dict[ch]
        else:
            str_builder += ch
    #print(str_builder)
    data_list_processed.append([d[0],str_builder])
    data_dict[d[0]].append(str_builder)
```

```python
data_list_processed = sorted(data_list_processed,key=lambda x:x[0])
with open('dictionary_extracted_processed.tsv','w',encoding='utf-8') as fop:
    for data in data_list_processed:
        fop.write(data[0]+"\t"+data[1]+"\n")
```
